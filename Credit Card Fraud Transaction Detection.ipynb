```python
def plot_confusion_matrix(cm, classes, normalize = False, title = "Confusion matrix", cmap = plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting normalize = True.
    Code from http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
    """
    plt.figure(dpi = 100)
    plt.imshow(cm, interpolation = 'nearest', cmap = cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype("float")/cm.sum(axis = 1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print("Confusion matrix, without normalization")

    #print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
                 horizontalalignment = "center",
                 color = "white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
def classification_report(true_y, predict_y, set_title = "Confusion matrix"):
    # construct confusion matrix
    conf_matrix = confusion_matrix(true_y, predict_y)
    # the accuracy under true label
    # True positive rate or Recall
    TPR = float(conf_matrix[1,1] / (conf_matrix[1,0] + conf_matrix[1,1]))
    # True negative rate
    TNR = float(conf_matrix[0,0] / (conf_matrix[0,0] + conf_matrix[0,1]))
    accuracy = accuracy_score(true_y, predict_y)
    
    with plt.style.context(("seaborn-white")):
        plot_confusion_matrix(conf_matrix, classes = ['Normal', 'Fraud'], title = set_title)
    
    print "\n------------------------------------Classification Report------------------------------------"
    # the accuracy for fraud given true proportion 
    print 'True positive rate is {}'.format(TPR)
    # the accuracy for normal given true prorortion
    print 'True negative rate is {}'.format(TNR)
    
```

### Undersampling

```python
# define features and classes
feature_name = list(data.columns[0:-1])
class_name = data.columns[-1]

np.random.seed(2)
# Get all X and y
X = np.array(data[feature_name])
y = np.array(data[class_name])
# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)

# have the training set
training_set = pd.DataFrame(X_train)
training_set['Class'] = y_train

# create a function to get downsampling data
def under_sampling(creditcard):
    sample_size = sum(creditcard.Class == 1)
    fraud_indices = creditcard[creditcard.Class == 1].index
    normal_indices =  creditcard[creditcard.Class == 0].index
    random_indices = np.random.choice(normal_indices, sample_size, replace=False) 
    merged_sample = np.concatenate([fraud_indices, random_indices])
    sample = creditcard.iloc[merged_sample]
    
    return sample

# Under sampling
under_sampled = under_sampling(training_set)
X_untrain = np.array(under_sampling(training_set).ix[:, 0:-1])
y_untrain = np.array(under_sampling(training_set).ix[:, -1])
```

# Model

## Logistic Classification

```python
def tuningLogitRegression(data_X, data_y, estimator, random_state):
    
    # A low C makes the decision surface smooth, while a high C aims at classifying all training examples 
    # correctly by giving the model freedom to select more samples as support vectors
    C_range = np.logspace(-3, 3, 7)  # Ridge penalty parameter.
    class_weight = [{1:2}]
    
    cv = KFold(n_splits = 5, random_state = random_state) # 5-fold cross validation.

    # Define a parameter dictionary. Parameters to be tuned: penalty, C, and fit_intercept in 
    # LogisticRegression.
    parameters = dict(penalty = ("l1", "l2"), C = C_range, class_weight = class_weight)
    
    find_best_est = GridSearchCV(estimator, param_grid = parameters, cv = cv, n_jobs = 4)
    find_best_est.fit(data_X, data_y)

    print("The best parameters are %s with a score of %0.2f" % (find_best_est.best_params_, 
                                                                find_best_est.best_score_))
    
    return find_best_est
```

### Undersampling

```python
# Use the function tuningLogitRegression to obtain the best parameters. 
un_log = tuningLogitRegression(X_untrain, y_untrain, LogisticRegression(), 4)
```

```python
un_log_pred = un_log.predict(X_test)
classification_report(y_test, un_log_pred, set_title = 'Confusion Matrix (Logistic) for Undersampling')
```

```python
print("The accuracy for using best parameters in logistic regression by under-sampling is {}."
      .format(accuracy_score(y_test.ravel(), un_log_pred)))
```

### Tuning parameters for Models

```python
def tuning_params_rbf(data_X, data_y, estimator):
    C_range = np.logspace(-3,3,7)  # ridge penalty parameter
    # A low C makes the decision surface smooth, while a high C aims at classifying all training examples correctly 
    # by giving the model freedom to select more samples as support vectors
    cv = KFold(n_splits=5, random_state = 5) # 5-fold cross validation

    gamma_range = np.logspace(-3,3,7) # the inverse of the radius of influence of samples selected by the model 
                                            # as support vectors
    param_grid = dict(gamma=gamma_range, C=C_range)
    grid = GridSearchCV(estimator, param_grid=param_grid, cv=cv, n_jobs = 4)
    grid.fit(data_X, data_y)

    print("The best parameters are %s with a score of %0.2f" % (grid.best_params_, grid.best_score_))
    return grid
```

```python
def tuning_params_linsvc(X, y, estimator):
    
    C_range = np.logspace(-3,3,7)
    cv = KFold(n_splits=5, random_state = 5)
    class_weight = [{1:3}]
    param_grid = dict(C=C_range, penalty = ("l1", "l2"), class_weight = class_weight)
    grid = GridSearchCV(estimator, param_grid=param_grid, cv=cv, n_jobs = 4)
    grid.fit(X, y)
    print("The best parameters are %s with a score of %0.2f" % (grid.best_params_, grid.best_score_))
    return grid
```

## Random Forest

```python
def tuningRandomForest(data_X, data_y, estimator, random_state):
    
    cv = KFold(n_splits = 5, random_state = random_state) # 5-fold cross validation.
    n_iter_search = 30
    
    # Define a parameter dictionary. Parameters to be tuned: n_estimators, criterion, max_features, 
    # and class_weight in RandomForest.
    parameters = dict(n_estimators = list(range(1,101)), 
                      criterion = ("gini", "entropy"), 
                      max_features = ("auto", "log2"),
                      min_samples_leaf = [1, 50, 100, 150],
                      max_depth = [3, 6, 8, 9, 10, 11, 12])
    
    
    find_best_est = RandomizedSearchCV(estimator, param_distributions = parameters, cv = cv, n_iter = n_iter_search,
                                      random_state = random_state)
    find_best_est.fit(data_X, data_y)

    print("The best parameters are %s with a score of %0.2f" % (find_best_est.best_params_, 
                                                                find_best_est.best_score_))
    
    return find_best_est
```

### Undersampling

```python
rf_param = tuningRandomForest(X_untrain, y_untrain, RandomForestClassifier(), 22)
```
